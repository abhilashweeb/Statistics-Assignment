#Theoritical Questions

1. What is hypothesis testing in statistics?
* Hypothesis testing is a method in statistics used to make decisions or inferences about a population based on sample data. It involves making an assumption (called the null hypothesis), then using data to test whether that assumption is likely to be true or should be rejected. If the data provides enough evidence, we reject the null hypothesis in favor of an alternative one.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
* The null hypothesis (H₀) is a statement that there is no effect or no difference—it assumes things are the same as usual. The alternative hypothesis (H₁ or Ha) is what we want to prove—it suggests there is a difference or effect. In hypothesis testing, we either reject the null or fail to reject it based on the evidence.

3. What is the significance level in hypothesis testing, and why is it important?
* The significance level is the threshold for deciding whether to reject the null hypothesis—commonly set at 0.05 (5%). It represents the probability of making a Type I error, which means rejecting the null hypothesis when it's actually true. It's important because it controls how confident we are in our test results.

4. What does a P-value represent in hypothesis testing?
* The P-value tells us the probability of getting our observed results (or more extreme) if the null hypothesis is actually true. A small P-value (typically less than 0.05) means the observed data is unlikely under the null hypothesis, so we may consider rejecting it. It helps us measure the strength of the evidence against the null.

5. How do you interpret the P-value in hypothesis testing?
* We interpret the P-value by comparing it to the significance level (α).
  * If the P-value is less than or equal to alpha, we reject the null hypothesis — there is strong evidence against it.
  * If the P-value is greater than alpha, we fail to reject the null hypothesis — the evidence is not strong enough.

6. What are Type 1 and Type 2 errors in hypothesis testing?
* A Type 1 error happens when we reject the null hypothesis even though it is actually true (a false positive).
* A Type 2 error occurs when we fail to reject the null hypothesis even though it is false (a false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
* In a one-tailed test, we check for an effect in only one direction (either greater than or less than a certain value). In a two-tailed test, we check for an effect in both directions (not equal to a certain value). The choice depends on what we are trying to prove with our hypothesis.

8. What is the Z-test, and when is it used in hypothesis testing?
* The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means) when the population standard deviation is known and the sample size is large (usually n > 30). We use it when the data is normally distributed and meets these conditions.

9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?
* We calculate the Z-score by subtracting the population mean from the sample Mean, then dividing the result by the population standard deviation divided by the square root of the sample size.
* The Z-score represents how many standard deviations our sample mean is away from the population mean. It helps us understand how unusual or typical our sample result is in the context of hypothesis testing.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
* The T-distribution is a probability distribution used in statistics when the sample size is small and the population standard deviation is unknown. We use It is used instead of the normal distribution when working with smaller samples (typically less than 30), because it accounts for the extra variability that comes with estimating the standard deviation from the sample.

11. What is the difference between a Z-test and a T-test?
* The Z-test is used when the sample size is large and the population standard deviation is known. The T-test is used when the sample size is small and the population standard deviation is unknown. Both tests help us compare means, but the T-test is better suited for smaller, more uncertain data.

12.  What is the T-test, and how is it used in hypothesis testing?
* The T-test is a statistical test used to compare the means of two groups or to compare a sample mean with a known value when the population standard deviation is unknown. We use it in hypothesis testing to determine if the difference between means is statistically significant or due to random chance, especially when dealing with small sample sizes.

13. What is the relationship between the Z-test and the T-test in hypothesis testing?
* The Z-test and T-test are both used in hypothesis testing to compare means. The main difference lies in the conditions: we use the Z-test when the population standard deviation is known and the sample size is large, while the T-test is used when the population standard deviation is unknown and the sample size is small. As the sample size increases, the T-distribution approaches the normal distribution, making both tests give similar results.

14. What is a confidence interval, and how is it used to interpret statistical results?
* A confidence interval is a range of values that we believe contains the true population parameter, based on our sample data, with a certain level of Confidence (like 95%). It helps us understand the precision of our estimate — a narrower interval means more precision. If the interval includes the hypothesised value, we usually do not reject the null hypothesis.

15. What is the margin of error, and how does it affect the confidence interval?
* The margin of error is the amount we add and subtract from the sample estimate to create a confidence interval. It reflects the uncertainty in our estimate — a larger margin of error means a wider confidence interval and less precision, while a smaller margin gives a narrower interval and more precision. It’s influenced by sample size and confidence level.

16. How is Bayes' Theorem used in statistics, and what is its significance?
* Bayes' Theorem is used in statistics to update the probability of an event based on new evidence. It helps us revise our beliefs by combining prior knowledge with current data. Its significance lies in making decisions or predictions in uncertain situations, especially in fields like medicine, finance, and machine learning.


17. What is the Chi-square distribution, and when is it used?
* The Chi-square distribution is a statistical distribution used to test relationships between categorical variables. We use it in tests like the Chi-square test for independence or the goodness-of-fit test to see if observed data matches expected data. It is commonly applied in surveys, genetics, and research involving frequency counts.

18. What is the Chi-square goodness of fit test, and how is it applied?
* The Chi-square goodness of fit test is used to check if the observed frequencies in a categorical dataset match the expected frequencies based on a specific hypothesis. We apply it by comparing the actual counts with the expected counts using a formula, then checking if the difference is statistically significant. It helps us see if the data fits a certain distribution.
19. What is the F-distribution, and when is it used in hypothesis testing?
* The F-distribution is a probability distribution used mainly to compare variances between two groups. In hypothesis testing, we use it in ANOVA (Analysis of Variance) and in tests comparing multiple means or variances. It helps us determine whether the differences between group variances or means are statistically significant.

20. What is an ANOVA test, and what are its assumptions?
* The ANOVA (Analysis of Variance) test is used to compare the means of three or more groups to see if at least one is significantly different. It checks whether the observed differences are likely due to chance or actual effects.
* Assumptions of ANOVA include:
   i. The data in each group is normally distributed.
  ii. The groups have equal variances (homogeneity of variance).
 iii. The observations are independent.

21. What are the different types of ANOVA tests?
* There are three main types of ANOVA tests, each used in different situations:
  1. One-Way ANOVA – Used to compare the means of three or more independent groups based on one factor.
  2. Two-Way ANOVA – Used when we want to study the effect of two different factors on a dependent variable.
  3. Repeated Measures ANOVA – Used when the same subjects are measured multiple times under different conditions or over time.

22. What is the F-test, and how does it relate to hypothesis testing?
* The F-test is a statistical test used to compare two variances or to test the Overall significance in ANOVA. It helps us determine whether the variability between group means is larger than what we'd expect by chance. In hypothesis testing, we use the F-test to see if there’s a significant difference between groups or models.

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

# Sample data
sample = [50, 55, 60, 54, 56, 40, 57, 55, 54]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)


pop_mean = 50


z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
print(f"Z-statistic: {z:.2f}")


p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}")


alpha = 0.05


if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm

# Sample data
data = [50, 55, 60, 54, 56, 40, 57, 55, 54]
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)


confidence = 0.95
z = norm.ppf(1 - (1 - confidence)/2)  # Z-score for 95% confidence ~1.96


margin_error = z * (std / np.sqrt(n))

lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"Sample Mean: {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [102, 98, 101, 100, 99, 103, 97, 100, 102, 101]

population_mean = 100
population_std = 2

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population means.")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [52, 50, 49, 51, 53, 47, 50, 52, 49]

population_mean = 50
population_std = 2

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = [52, 50, 49, 51, 53, 47, 50, 52, 49]

population_mean = 50
population_std = 2

sample_mean = np.mean(data)
n = len(data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

print("Z-score:", round(z_score, 2))
print("P-value:", round(p_value, 4))

if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# Plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Normal Curve')
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z = {z_score:.2f}')

plt.title('Two-Tailed Z-Test')
plt.xlabel('Z value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualize Type I and Type II errors for hypothesis testing.

    Parameters:
    - mu0: mean under null hypothesis
    - mu1: mean under alternative hypothesis
    - sigma: population standard deviation
    - n: sample size
    - alpha: significance level (default = 0.05)
    """
    se = sigma / np.sqrt(n)

    # Critical z-value for two-tailed test
    z_crit = norm.ppf(1 - alpha / 2)

    # Critical sample means
    crit_left = mu0 - z_crit * se
    crit_right = mu0 + z_crit * se

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H₁)', color='green')

    plt.fill_between(x, y_null, where=(x < crit_left) | (x > crit_right),
                     color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, y_alt, where=(x >= crit_left) & (x <= crit_right),

                     color='orange', alpha=0.3, label='Type II Error (β)')

    plt.axvline(crit_left, color='black', linestyle='--', label='Critical Values')
    plt.axvline(crit_right, color='black', linestyle='--')

    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
plot_type1_type2_errors(mu0=50, mu1=52, sigma=5, n=30)
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy.stats import ttest_ind

# Sample data from two independent groups
group1 = [20, 25, 19, 24, 30, 28, 27]
group2 = [10, 20, 17, 22, 19, 21, 20]

t_statistic, p_value = ttest_ind(group1, group2)

print("T-statistic:", round(t_statistic, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The group means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the group means.")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - (lower bound, upper bound) of the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample standard error
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return round(lower, 2), round(upper, 2)

# Example usage
sample_data = [50, 52, 47, 51, 53, 49, 54, 48, 50, 52]
ci = confidence_interval(sample_data)
print("95% Confidence Interval for the Sample Mean:", ci)
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.75):
    """
    Calculates the margin of error for the sample mean at a given confidence level.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.75 for 91%)

    Returns:
    - margin of error (rounded to 2 decimal places)
    """
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample standard error
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    moe = t_critical * std_err
    return round(moe, 2)

# Example usage
sample_data = [100, 102, 98, 105, 101, 99, 97, 103, 100, 104]
moe = margin_of_error(sample_data, confidence=0.91)

print("Margin of Error at 91% Confidence Level:", moe)
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, likelihood, false_positive, base_rate):
    """
    Apply Bayes' Theorem to calculate the posterior probability.

    Parameters:
    - prior: P(H) = prior probability (e.g., disease rate)
    - likelihood: P(D|H) = probability of testing positive given disease
    - false_positive: P(D|¬H) = probability of testing positive without disease
    - base_rate: P(D) = total probability of testing positive (computed internally)

    Returns:
    - Posterior probability: P(H|D)
    """
    # Total probability of positive test (evidence)
    p_disease = prior
    p_no_disease = 1 - prior

    p_positive = (likelihood * p_disease) + (false_positive * p_no_disease)

    # Bayes' Theorem
    posterior = (likelihood * p_disease) / p_positive
    return round(posterior, 4)

# Example values
prior = 0.01             # P(H): 1% of people have the disease
likelihood = 0.99        # P(Pos|Disease)
false_positive = 0.05    # P(Pos|No Disease)

posterior = bayes_theorem(prior, likelihood, false_positive, base_rate=None)
print("Probability that the person actually has the disease:", posterior)
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Rows: Gender, Columns: Product Preference
data = pd.DataFrame({
    'Product A': [30, 50],  # Males, Females
    'Product B': [10, 40]
}, index=['Male', 'Female'])

chi2, p, dof, expected = chi2_contingency(data)

print("Observed Data:\n", data)
print("\nExpected Frequencies:\n", pd.DataFrame(expected, index=data.index, columns=data.columns))
print("\nChi-square Statistic:", round(chi2, 3))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nReject the null hypothesis: There is a significant association between gender and product preference.")
else:
    print("\nFail to reject the null hypothesis: No significant association between gender and product preference.")
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Observed frequencies from an experiment (e.g., dice rolls)
observed = np.array([8, 9, 10, 12, 11, 10])

# Expected frequencies (e.g., uniform distribution for a fair die)
expected = np.array([10, 10, 10, 10, 10, 10])

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", round(chi_stat, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fits the expected distribution well.")
```
