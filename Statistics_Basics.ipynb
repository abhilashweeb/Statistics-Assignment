#Theoritical Questions

1. What is statistics, and why is it important?


*   Statistics is the science of collecting, analysing, interpreting, and presenting data. It helps us make sense of large amounts of information. We use statistics to make informed decisions, find patterns, and solve real-world problems in areas like business, health, and research.


2. What are the two main types of statistics?
* The two main types of statistics are descriptive and inferential.
* Descriptive statistics help us summarize and organize data using charts, averages, or percentages.
* Inferential statistics allow us to make predictions or draw conclusions about a larger group based on a sample of data.

3. What are descriptive statistics?
* Descriptive statistics are methods that help us summarize, organize, and simplify data. We use measures like mean, median, mode, range, and standard deviation, along with charts and tables, to understand what the data is showing without making predictions.

4. What is inferential statistics?
* Inferential statistics help us make predictions or draw conclusions about a larger population based on data from a sample. We use techniques like hypothesis testing, confidence intervals, and regression analysis to make informed guesses and decisions.

5. What is sampling in statistics?
* Sampling in statistics is the process of selecting a smaller group (sample) from a larger group (population) to study. We use samples because it's often not possible or practical to study an entire population, and a well-chosen sample gives us insights about the whole.

6. What are the different types of sampling methods?
* There are several types of sampling methods, mainly divided into probability and non-probability sampling.
* In Probability Sampling
  * Simple Random Sampling
  * Stratified Sampling
  * Cluster Sampling
  * Systematic Sampling
* In Non-Probability Sampling
  * Convenience Sampling
  * Judgmental

7. What is the difference between random and non-random sampling?
* In random sampling, every member of the population has an equal chance of being selected, which reduces bias. In non-random sampling, we select samples based on factors like convenience or judgment, which may introduce bias and Affect the accuracy of our conclusions.

8. Define and give examples of qualitative and quantitative data.
* Qualitative data describes qualities or categories and is non-numerical, like colours, names, or types (e.g., eye colour, gender, or car brand).
* Quantitative data is numerical and measures quantities — like height, age, or salary (e.g., 160 cm tall, 25 years old, or ₹50,000 salary).

9. What are the different types of data in statistics?
* In statistics, data is mainly classified into two types:
* Qualitative (categorical) and Quantitative (numerical). Quantitative data is further divided into discrete (such as the number of students) and continuous (such as height or weight). Qualitative data includes nominal (such as colours) and ordinal (like rankings).

10. Explain nominal, ordinal, interval, and ratio levels of measurement.
* The four levels of measurement are:
  * Nominal: Labels or categories with no order (e.g., colors, gender).
  * Ordinal: Categories with a meaningful order, but no fixed gaps (e.g., ranks, ratings)
  * Interval: Numeric data with equal intervals but no true zero (e.g., temperature in Celsius).
  * Ratio: Numeric data with equal intervals and a true zero (e.g., weight, age, income).


11. What is the measure of central tendency?
* The measure of central tendency helps us find the center or average of a dataset. The main types are mean (average), median (middle value), and mode (most frequent value). We use them to understand where most data points tend to cluster.

12. Define mean, median, and mode.
* Mean is the average, found by adding all values and dividing by the number of values.
* Median is the middle value when the data is arranged in order.
* Mode is the value that appears most often in the dataset.

13. What is the significance of the measure of central tendency?
* The measure of central tendency helps us understand the general or typical value in a dataset. It gives us a single value that represents the whole set, making it easier to compare and interpret data in real-life situations like test scores, salaries, or survey results.

14. What is variance, and how is it calculated?
* Variance tells us how much the data values spread out from the mean. We calculate it by finding the difference of each value from the mean, squaring those differences, adding them up, and dividing by the number of values (or one less, if it's a sample). It helps us measure consistency in data.

15. What is standard deviation, and why is it important?
* Standard deviation measures how much the values in a dataset vary or spread out from the mean. It is the square root of the variance. It’s important because it tells us how consistent or dispersed our data is — smaller values mean data points are close to the mean, and larger values show more spread.

16. Define and explain the term range in statistics.
* In statistics, range is the difference between the highest and lowest values in a dataset. We calculate it by subtracting the smallest value from the largest. It gives us a quick idea of how spread out the data is actually.

17. What is the difference between variance and standard deviation?
* Variance and standard deviation both measure the spread of data, but in their way. Variance is the average of the squared differences from the mean, while standard deviation is the square root of the variance. Standard deviation is more commonly used because it's in the same units as the original data.

18. What is skewness in a dataset?
* Skewness tells us about the asymmetry of a dataset's distribution. If data is symmetrical, skewness is zero. A positive skew means most values are on the Left with a long right tail, and a negative skew means most values are on the right with a long left tail. It helps us understand the shape of the data.



19. What does it mean if a dataset is positively or negatively skewed?
* If a dataset is positively skewed, most values are concentrated on the left, and the tail stretches to the right — the mean is greater than the median. If it’s negatively skewed, most values are on the right, and the tail extends to the left — the mean is less than the median. This shows us how the data leans.

20. Define and explain kurtosis.
* Kurtosis measures the "tailedness" or peakedness of a data distribution. It tells us how much of the variance is due to extreme values (outliers).
* High kurtosis means sharp peaks and heavy tails.
* Low kurtosis means flat peaks and light tails.

21. What is the purpose of covariance?
* Covariance helps us understand how two variables change together. If both variables increase or decrease together, covariance is positive. If one Increases while the other decreases, it’s negative. It shows the direction of the relationship, but not the strength.

22.  What does correlation measure in statistics?
* The Correlation measures the strength and direction of the relationship between two variables. A positive correlation means both increase together, a negative one means one increases while the other decreases. We often use the correlation coefficient (r), which ranges from -1 to +1, to express this.

23. What is the difference between covariance and correlation?
* Covariance shows the direction of how two variables change together, but its value depends on the units of the data.
* On the other hand, Correlation also shows the direction, but it goes further by measuring the strength of the relationship on a standardised scale from -1 to +1. Correlation is easier to compare across datasets.


24. What are some real-world applications of statistics?
* Statistics is used in many real-world areas to make informed decisions. In healthcare, it helps track disease trends; in business, it guides marketing and Sales strategies; in sports, it analyses player performance; and in government, it shapes public policies using population data. We rely on it to interpret data and solve problems effectively.

```python
# 12. Implement and compare simple random sampling and systematic sampling.
import pandas as pd
import numpy as np


data = pd.DataFrame({'ID': range(1, 31)})
print("Original Dataset:")
print(data)
```

```python
# Simple Random Sampling: randomly pick 5 rows
random_sample = data.sample(n=5, random_state=52)

print("\nSimple Random Sample:")
print(random_sample)
```

```python
def systematic_sampling(df, sample_size):
    step = len(df) // sample_size
    indices = np.arange(0, len(df), step)
    return df.iloc[indices[:sample_size]]

systematic_sample = systematic_sampling(data, 5)

print("\nSystematic Sample:")
print(systematic_sample.T)
```

```python
# 13. Calculate the mean, median, and mode of grouped data.# 14. Simulate data using Python and calculate its central tendency and dispersion.
import numpy as np
import pandas as pd
from scipy.stats import mode

scores = [
    45, 48, 50, 52, 55, 56, 58, 60, 61, 62, 63, 65, 66, 67, 68,
    70, 72, 74, 75, 77, 80, 82, 85, 87, 90, 90, 90, 100, 105, 110
]

df = pd.DataFrame({'Scores': scores})

mean_val = df['Scores'].mean()
median_val = df['Scores'].median()
mode_val = mode(df['Scores'], keepdims=True).mode[0]

data_range = df['Scores'].max() - df['Scores'].min()
variance_val = df['Scores'].var()
std_dev_val = df['Scores'].std()

print("Central Tendency:")
print("Mean:", round(mean_val, 2))
print("Median:", round(median_val, 2))
print("Mode:", mode_val)

print("\nDispersion:")
print("Range:", round(data_range, 2))
print("Variance:", round(variance_val, 2))
print("Standard Deviation:", round(std_dev_val, 2))

# Class intervals and frequencies
classes = ['0-100', '100-200', '200-300', '300-400', '400-500']
frequencies = [5, 10, 15, 20, 25]

midpoints = []
for interval in classes:
    lower, upper = map(int, interval.split('-'))
    midpoints.append((lower + upper) / 2)

fx = [frequencies[i] * midpoints[i] for i in range(len(frequencies))]
mean = sum(fx) / sum(frequencies)

cumulative_frequencies = [sum(frequencies[:i+1]) for i in range(len(frequencies))]
N = sum(frequencies)
half_N = N / 2

for i in range(len(cumulative_frequencies)):
    if cumulative_frequencies[i] >= half_N:
        median_class_index = i
        break

L = int(classes[median_class_index].split('-')[0])
F = cumulative_frequencies[median_class_index - 1] if median_class_index != 0 else 0
f = frequencies[median_class_index]
h = 10

median = L + ((half_N - F) / f) * h

modal_index = frequencies.index(max(frequencies))
L_modal = int(classes[modal_index].split('-')[0])
f1 = frequencies[modal_index]
f0 = frequencies[modal_index - 1] if modal_index > 0 else 0
f2 = frequencies[modal_index + 1] if modal_index < len(frequencies) - 1 else 0

mode = L_modal + ((f1 - f0) / ((2 * f1) - f0 - f2)) * h

print("Grouped Data:")
print("Mean:", round(mean, 2))
print("Median:", round(median, 2))
print("Mode:", round(mode, 2))
```

```python
# 18. Implement Z-score normalization and explain its significance.
import numpy as np

data = [30, 48, 50, 52, 55, 56, 58, 60, 61, 62, 63, 65, 66, 67, 68, 70, 72, 74, 75, 77, 80, 82, 85, 87, 90, 100]

data_array = np.array(data)

mean = np.mean(data_array)
std = np.std(data_array)

z_scores = (data_array - mean) / std


print("Original Data:\n", data_array)
print("\nZ-score Normalized Data:\n", np.round(z_scores, 2))
```

```python
# 20.Write a Python program to visualize covariance using a heatmap.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math_Score': [35, 75, 80, 85, 90, 95, 100],
    'Science_Score': [60, 77, 82, 87, 92, 97, 95],
    'English_Score': [50, 68, 70, 72, 74, 76, 78]
}

df = pd.DataFrame(data)

cov_matrix = df.cov()

plt.figure(figsize=(6, 4))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Covariance Heatmap")
plt.tight_layout()
plt.show()
```

```python
# 24. Implement the Pearson and Spearman correlation coefficients for a dataset.
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Exam_Score':    [30, 40, 50, 55, 60, 65, 70, 75, 85]
}

df = pd.DataFrame(data)

pearson_corr, _ = pearsonr(df['Hours_Studied'], df['Exam_Score'])

spearman_corr, _ = spearmanr(df['Hours_Studied'], df['Exam_Score'])

print("Pearson Correlation Coefficient:", round(pearson_corr, 2))
print("Spearman Correlation Coefficient:", round(spearman_corr, 2))
```
